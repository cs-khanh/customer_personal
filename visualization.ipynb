```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the uploaded CSV file
df = pd.read_csv("customer_segmentation_clean.csv")

# Summary statistics
summary = df.groupby('Age_Group').agg({
    'Income': 'mean',
    'Expenses': 'mean',
    'TotalNumPurchases': 'sum'
}).reset_index()

print(summary)
```

```python
# Pie chart for marital status distribution
marital_status = df.groupby(['Age_Group', 'Marital_Status']).size().unstack().fillna(0)
marital_status.plot(kind='pie', subplots=True, figsize=(18, 6), autopct='%1.1f%%', legend=False)
for ax in plt.gcf().axes:
    ax.set_title(ax.get_title(), fontsize=12)
plt.suptitle('Marital Status Distribution by Age Group')
plt.show()
```

```python
# Scatter plot for income vs expenses using seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='Expenses', hue='Age_Group')
plt.title('Income vs Expenses')
plt.xlabel('Income')
plt.ylabel('Expenses')
plt.show()
```
