```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customer_segmentation_clean.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age_Group', y='Expenses', palette='Set3')
plt.title('Phân phối chi tiêu theo nhóm tuổi')
plt.xlabel('Nhóm tuổi')
plt.ylabel('Chi tiêu')
plt.xticks(rotation=45)
plt.show()
```

```python
# Nhận xét về biểu đồ hộp (Boxplot) phân phối chi tiêu theo nhóm tuổi:
#  -Nhóm tuổi "Middle-Aged Adults" có nhiều khách hàng tiềm năng chi tiêu ít hơn so với các nhóm tuổi khác, nhưng lại có một số ít người chi tiêu rất cao, thể hiện qua các outliers.
#  -Nhóm "Elders" và "Young Adults" có phạm vi chi tiêu rộng, nhưng nhóm "Elders" có sự phân tán chi tiêu lớn hơn, thể hiện qua hộp rộng và whiskers dài.
#  -Nhóm tuổi nào có tiềm năng chi tiêu cao nhất?: Nhóm "Elders" có thể là một nhóm tuổi tiềm năng vì có nhiều khách hàng chi tiêu cao hơn trung bình.
```
